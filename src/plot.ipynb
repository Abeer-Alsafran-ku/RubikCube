# Plot Brute-force and Proposed 

```python
import matplotlib.pyplot as plt
import numpy as np
import math
# Sizes to test
sizes = np.array([ 1, 2, 5, 10,12,15])

# Time functions
def time_kociemba(s):
    t = []
    for size in s:  # Iterate over each size
        t.append(size)  # Add size directly to the list
    return t

def time_bruteforce(s):
    t = []
    for size in s:  # Iterate over each size
        # t.append(2 ** (4 * size - 1))  # Compute time and append to list
        t.append(pow(2 ,4*size - 1))  # Compute time and append to list
    return t

# Compute times
kociemba_times = time_kociemba(sizes)
bruteforce_times = time_bruteforce(sizes)

print("Size",sizes)
print("Brute-Force",bruteforce_times)
print("Modified Kociemba",kociemba_times)


# Plotting
plt.figure(figsize=(10, 6))
plt.plot(sizes, kociemba_times, label='Proposed Modified Kociemba', marker='o')
plt.plot(sizes, bruteforce_times, label='Brute-force', marker='x')

plt.yscale('symlog')  
plt.xlabel('Size (s)')
plt.ylabel('Time (log scale)')
plt.title('Time vs Size: Brute-force vs Modified Kociemba')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
